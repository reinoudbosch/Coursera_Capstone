import postalcode wikipedia page containing postal codes of Amsterdam

```python
import requests
from bs4 import BeautifulSoup
source = requests.get('https://nl.wikipedia.org/wiki/Lijst_van_postcodes_1000-1999_in_Nederland#1000-1099').text
soup = BeautifulSoup(source,'lxml')
```

The wikipedia page indicates that the relevant postalcodes range from 1000 to 1099 (excluding the district of South-East which is separated from the rest of the municipality of Amsterdam by another municipality). Looking at the source of the page, the relevant information is contained within li elements within an ul element within the second ul element. The boroughlist containing boroughs and neighborhoods is extracted from beautifulsoup object soup.

```python
uls=soup.find_all('ul')
sub_ul=uls[1].find('ul')
boroughlist = []
lis=sub_ul.find_all('li', recursive=False)
for elem in lis:
    boroughlist.append(elem.text.strip())
boroughlist
```

From this list, the boroughs and neighborhoods can be extracted. 

```python
elements = []
for element in boroughlist:
    temp=element.split('\n')
    elements.append(temp)
elements
```

```python
b_and_ns = []
for borough in elements:
    try:
        borough[1]
    except:         # the borough does not contain neighborhoods
        b_and_ns.append((borough[0], borough[0]))
        continue
    for neighborhood in borough[1:]:   # borough contains neighborhoods
        b_and_ns.append((borough[0], neighborhood))
b_and_ns
```

Insert list into dataframe.

```python
import pandas as pd
headers = ['Borough','Neighborhood']
amsterdam_data=pd.DataFrame(b_and_ns, columns=headers)
amsterdam_data=amsterdam_data.drop(61) # drop row 'vervallen' (deprecated)
amsterdam_data
```

Strip postal code away from borough column

```python
amsterdam_data['Borough'] = amsterdam_data['Borough'].map(lambda x: x.lstrip('0123456789-: '))
amsterdam_data
```

Retrieve postal code from neighborhood column and store in column PostalCode

```python
amsterdam_data['PostalCode'] = amsterdam_data['Neighborhood'].astype(str).str[:4]
amsterdam_data['Neighborhood'] = amsterdam_data['Neighborhood'].map(lambda x: x.lstrip('0123456789-: '))
columnsTitles=['PostalCode','Borough','Neighborhood']
amsterdam_data=amsterdam_data.reindex(columns=columnsTitles)
amsterdam_data
```

retrieve latitudes and longitudes of postalcodes from arcgis

```python
import geocoder
import json
coordinates = []
for i in range(0, len(amsterdam_data)):
    postal_code = amsterdam_data.iloc[i]['PostalCode']
    string = '{} Amsterdam, NL'.format(postal_code)
    g = geocoder.arcgis(string)
    coordinates.append((postal_code,g.json['lat'],g.json['lng']))
coordinates
```

add coordinates to dataframe


```python
headers = ['PostalCode','Latitude','Longitude']
lat_lng_data = pd.DataFrame(coordinates, columns=headers)
amsterdam_data = pd.merge(amsterdam_data, lat_lng_data, on='PostalCode', sort=False)
amsterdam_data
```

Define function to explore nearby venues in boroughs

```python
LIMIT=100
radius=500
def getNearbyVenues(names, latitudes, longitudes, radius=500):
    
    venues_list=[]
    for name, lat, lng in zip(names, latitudes, longitudes):
            
        # create the API request URL
        url = 'https://api.foursquare.com/v2/venues/explore?&client_id={}&client_secret={}&v={}&ll={},{}&radius={}&limit={}'.format(
            CLIENT_ID, 
            CLIENT_SECRET, 
            VERSION, 
            lat, 
            lng, 
            radius, 
            LIMIT)
            
        # make the GET request
        results = requests.get(url).json()["response"]['groups'][0]['items']
        
        # return only relevant information for each nearby venue
        venues_list.append([(
            name, 
            lat, 
            lng, 
            v['venue']['name'], 
            v['venue']['location']['lat'], 
            v['venue']['location']['lng'],  
            v['venue']['categories'][0]['name']) for v in results])

    nearby_venues = pd.DataFrame([item for venue_list in venues_list for item in venue_list])
    nearby_venues.columns = ['Borough',
                  'Borough Latitude', 
                  'Borough Longitude', 
                  'Venue', 
                  'Venue Latitude', 
                  'Venue Longitude', 
                  'Venue Category']
    
    return(nearby_venues)
```

First use requests and BeautifulSoup to read the html of the Toronto postal codes into variable soup

```python
source = requests.get('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
soup = BeautifulSoup(source,'lxml')
```

Use BeautifulSoup to read the html of the postalcode table into variable table. 
Then read the body of the table into variable table_body. 
Next, read the table rows into variable table_rows.
Then iterate through the table rows to read the text of the table elements (td) into the variable records. 
Display the contents of variable records

```python
table=soup.find('table', class_='wikitable sortable')
table_body=table.find('tbody')
table_rows=table_body.find_all('tr')

del table_rows[0]         # delete empty row coming from header

records = []
for row in table_rows:
    cells = row.find_all('td')
    cells=[cell.text.strip() for cell in cells]   # find text inside td
    records.append(cells)

records
```

Set up a pandas dataframe variable toronto_data with column headers 'PostalCode', 'Borough', and 'Neighborhood'. Fill the dataframe with the contents of records. Where the value of 'Neighborhood' is 'Not assigned' change that value to the value of 'Borough'. Group by 'PostalCode' and 'Borough' and combine 'Neighborhood' values separated by commas. Display the dataframe.

```python
headers = ['PostalCode','Borough','Neighborhood']
toronto_data = pd.DataFrame(records, columns=headers)
toronto_data = toronto_data[toronto_data['Borough'] != 'Not assigned']

# where neighborhood == 'Not assigned' make Neighborhood = Borough
cols = toronto_data.columns[toronto_data.columns.str.contains('Neighborhood')]
toronto_data[cols] = toronto_data[cols].mask(toronto_data[cols].apply(lambda x: x.str.contains('Not assigned')), toronto_data['Borough'], axis=0)

toronto_data = toronto_data.groupby(['PostalCode','Borough'], sort=False).agg(lambda x: ', '.join(x)).reset_index()
toronto_data
```

Confirm the new size

```python
amsterdam_grouped.shape
```

Print each borough along with the top 5 most common venues

```python
num_top_venues = 5

for borough in amsterdam_grouped['Borough']:
    print("----"+borough+"----")
    temp = amsterdam_grouped[amsterdam_grouped['Borough'] == borough].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

function to sort venues in descending order

```python
def return_most_common_venues(row, num_top_venues):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:num_top_venues]
```

create new dataframe and display top 10 venues for each borough

```python
import numpy as np
num_top_venues = 10

indicators = ['st', 'nd', 'rd']

# create columns according to number of top venues
columns = ['Borough']
for ind in np.arange(num_top_venues):
    try:
        columns.append('{}{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{}th Most Common Venue'.format(ind+1))

# create a new dataframe
boroughs_venues_sorted = pd.DataFrame(columns=columns)
boroughs_venues_sorted['Borough'] = amsterdam_grouped['Borough']

for ind in np.arange(amsterdam_grouped.shape[0]):
    boroughs_venues_sorted.iloc[ind, 1:] = return_most_common_venues(amsterdam_grouped.iloc[ind, :], num_top_venues)

boroughs_venues_sorted
```

Run k-means to cluster boroughs

```python
# !conda install -c conda-forge scikit-learn --yes   # uncomment this if scikit-learn is not yet installed

from sklearn.cluster import KMeans
# set number of clusters
kclusters = 5

amsterdam_grouped_clustering = amsterdam_grouped.drop('Borough', 1)

# run k-means clustering
kmeans = KMeans(n_clusters=kclusters, random_state=0).fit(amsterdam_grouped_clustering)

# check cluster labels generated for each row in the dataframe
kmeans.labels_[0:10] 
```

create new dataframe that includes the cluster as well as the top 10 venues for each borough

```python
# add clustering labels
boroughs_venues_sorted.insert(0, 'Cluster Labels', kmeans.labels_)
```

```python
amsterdam_merged = amsterdam_data

# merge amsterdam_grouped with amsterdam_data to add latitude/longitude for each neighborhood
amsterdam_merged = amsterdam_merged.join(boroughs_venues_sorted.set_index('Borough'), on='Borough')

amsterdam_merged.head()
```

Now that we have five clusters for both Amsterdam and Toronto, the harmonies between the different clusters can be analyzed. For this purpose, Euclidean distance will be calculated pairwise between the Amsterdam and Toronto clusters resulting in an Euclidean distance matrix. This distance matrix will be sorted retaining the indexes. Then the Euclidean distance scores between the different cluster centers will be printed in ascending order.

```python
import scipy.spatial.distance as dist
import numpy as np

edm = np.zeros((5,5))
for i in range(0, 5):
    for j in range(0, 5):
        edm[i][j]=dist.euclidean(amsterdam_cluster_centers[i], toronto_cluster_centers[j])

idx=np.argsort(edm, axis=None)     
rows, cols = np.unravel_index(idx, edm.shape)
edm_sorted = edm[rows, cols]
for r, c, v in zip(rows, cols, edm_sorted):
     print('amsterdam_cluster_center['+'01234'[r]+']', '&', 'toronto_cluster_center['+'01234'[c]+']', ':', v)
```
